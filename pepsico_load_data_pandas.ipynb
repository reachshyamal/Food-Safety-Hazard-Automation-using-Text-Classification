```python
col=['Filename','Allergen_text','Label_text','Tag']
df1=pd.DataFrame(columns=col)
dir1='D:\\Pepsico'
for file_path in glob.glob(os.path.join(dir1, '*.xlsx')):
    print(file_path)
    with open(file_path, 'r') as f:
        data1=pd.read_excel(file_path,sheetname=xls.sheet_names()[1])
        data2=data1[data1['Version:'].apply(lambda x: str(x).__contains__('Allergens'))]['5.4.1'].values
        data3=data1[data1['Version:'].apply(lambda x: str(x).__contains__('Labeling Instructions:  '))]['5.4.1'].values
        data4=data1[data1['Version:'].apply(lambda x: str(x).__contains__('Process platform '))]['Unnamed: 4'].values.flatten()
        #values=[[os.path.basename(file_path),data2,data3,random.choices(['Y','N'],weights=[0.5,0.5])]]
        values=[[os.path.basename(file_path),data2,data3,data4]]
        data_frame = pd.DataFrame(values, columns=col)
        df1=df1.append(data_frame,ignore_index=True)
          
```
